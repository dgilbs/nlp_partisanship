```python
def get_date2(date_str):
    date_arr = date_str.split(",")
    year = int(date_arr[1])
    month_name = date_arr[0].split(" ")[0].title()
    day = int(date_arr[0].split(" ")[1])
    months = list()
    for num in range(1,13):
        months.append(calendar.month_name[num])
    months_dict = dict(zip(months, range(1,13)))
    month = months_dict[month_name]
    return datetime(year=year, month=month, day=day)
    
get_date2("March 26, 2017")
```

```python
def get_date3(date_str): # for when the day of the week is in the name
    date_arr = date_str.split(", ")[1:]
    #print(date_arr)
    year = int(date_arr[1])
    month_name = date_arr[0].split(" ")[0].title()
    day = int(date_arr[0].split(" ")[1])
    months = list()
    for num in range(1,13):
        months.append(calendar.month_name[num])
    months_dict = dict(zip(months, range(1,13)))
    month = months_dict[month_name]
    return datetime(year=year, month=month, day=day)

get_date3("Monday, January 5, 2009")
```

```python
def adjust_df4(committee):
    maj_file = "/Users/danielgilberg/data_science/metis/nlp_partisanship/data/{}/majority.pkl".format(committee)
    min_file = "/Users/danielgilberg/data_science/metis/nlp_partisanship/data/{}/minority.pkl".format(committee)
    maj_df = pd.read_pickle(maj_file)
    min_df = pd.read_pickle(min_file)
    maj_df["Date"] = maj_df.apply(lambda row: get_date2(row["Date"]), axis=1)
    min_df["Date"] = min_df.apply(lambda row: get_date2(row["Date"]), axis=1)
    turnover_date = datetime(year=2015, month=1, day=3)
    dem_maj_df = maj_df[maj_df.Date < turnover_date]
    dem_min_df = min_df[min_df.Date >= turnover_date]
    rep_maj_df = maj_df[maj_df.Date >= turnover_date]
    rep_min_df = min_df[min_df.Date < turnover_date]
    rep_arr = [rep_maj_df, rep_min_df]
    dem_arr = [dem_min_df, dem_maj_df]
    rep_df = pd.concat(rep_arr)
    dem_df = pd.concat(dem_arr)
    rep_df["Party"] = "Republican"
    rep_df["Committee"] = committee.replace("_", " ").title()
    rep_df["Label"] = "Republican, " + committee.replace("_", " ").title()
    dem_df["Party"] = "Democratic"
    dem_df["Committee"] = committee.replace("_", " ").title()
    dem_df["Label"] = "Democratic, " + committee.replace("_", " ").title()
    dem_file = "/Users/danielgilberg/data_science/metis/nlp_partisanship/data/{}/democratic.pkl".format(committee)
    with open(dem_file, "wb") as picklefile:
        pickle.dump(dem_df, picklefile)
    rep_file = "/Users/danielgilberg/data_science/metis/nlp_partisanship/data/{}/republican.pkl".format(committee)
    with open(rep_file, "wb") as picklefile:
        pickle.dump(rep_df, picklefile)
    return rep_df
```
