```python
#scraping the majority press links for the Appropriations committee
def get_maj_appr_links():
    appr_maj_links = list()
    url = "https://www.appropriations.senate.gov/news/majority"
    driver.get(url)
    form = driver.find_element_by_class_name("pagination-right")
    count = int(form.text.split("of")[1])
    #https://www.appropriations.senate.gov/news/majority?PageNum_rs=2&
    for num in range(1, count+1):
        temp = "https://www.appropriations.senate.gov/news/majority?PageNum_rs={}&".format(str(num))
        driver.get(temp)
        arr = driver.find_elements_by_class_name("title")
        for i in arr:
            appr_maj_links.append(i.find_element_by_tag_name("a").get_attribute("href"))
    return appr_maj_links

appr_maj_links = get_maj_appr_links()
```

```python
def get_banking_links(party):
    url = "https://www.banking.senate.gov/public/index.cfm/{}-press-releases".format(party.lower())
    driver.get(url)
    arr = driver.find_elements_by_class_name("recordListTitle")
    button = driver.find_elements_by_class_name("navbar-inner")
    page_count = int(button[1].text.split("First")[0].split("of")[1].split("\n")[0])
    
    final = list()
    for num in range(1, page_count+1):
        #print(num)
        url = "https://www.banking.senate.gov/public/index.cfm/{}-press-releases?page={}".format(party, str(num))
        driver.get(url)
        arr = driver.find_elements_by_class_name("recordListTitle")
        for i in arr:
            link = i.find_element_by_tag_name("a")
            final.append(link.get_attribute("href"))
    return final
# gop_banking_links = get_banking_links("republican")
# dem_banking_links = get_banking_links("democratic")
```

```python
def get_banking_releases(party):
    links = get_banking_links(party)
    rows = list()
    print("links are done")
    print(len(links))
    for index, url in enumerate(links):
        print (index)
        driver.get(url)
        try:
            title = driver.find_element_by_class_name("title")
            title = title.text
        except:
            title = ""
        try:
            date = driver.find_element_by_class_name("date")
            date = date.text
        except:
            date = ""
        try:
            content = driver.find_element_by_class_name("content")
            release = content.text.replace("\n", "")
        except:
            release = ""
        row = [date, title, release]
        rows.append(row)
    df = pd.DataFrame(rows, columns=["Date", "Title", "Release"])
    file_path = "/Users/danielgilberg/data_science/metis/nlp_partisanship/data/banking/{}.pkl".format(party)
    with open(file_path, "wb") as picklefile:
        pickle.dump(df, picklefile)
    return df

get_banking_releases("democratic")
```
