## Elizabeth Warren <a class="anchor" id="warren"></a>

```python
warren_links = list()
for num in range(1,100):
    print(num)
    url = "https://www.warren.senate.gov/index.cfm?p=press_releases&pg={}".format(num)
    driver.get(url)
    titles = driver.find_elements_by_class_name("title")

    for t in titles:

        try:
            link = t.find_element_by_tag_name("a")
            if("press_release" in link.get_attribute("href")):
                warren_links.append(link.get_attribute("href"))
        except:
            pass
```

```python
def get_warren_releases(arr):
    rows = list()
    for index, url in enumerate(arr):
        print(index)
        driver.get(url)
        try:
            headline = driver.find_element_by_class_name('page-title').text
        except:
            headline = ""
            
        try:
            date = driver.find_element_by_class_name("inner-lyt-dates")
            date = date.text.title()
        except:
            date = ""
        try:
            release = driver.find_element_by_class_name("record-bd").text
        except:
            release = ""
            
        row = [date, headline, release]
        rows.append(row)
    df = pd.DataFrame(rows, columns=["Date", "Title", "Release"])
    return df


warren_df = get_warren_releases(warren_links)
    
```

```python
warren_df["Senator"] = "Elizabeth Warren"
```

```python
def get_warren_date(date_str):
    try:
        date_arr = date_str.split(", ")
        year= int(date_arr[1])
        mo = date_arr[0].split(" ")[0]
        d = dict((v,k) for k,v in enumerate(calendar.month_abbr))
        month = d[mo]
        day = int(date_arr[0].split(" ")[1])
        return datetime(year=year, month=month, day=day)
    except:
        return ''

warren_df
warren_df["Date"] = warren_df.apply(lambda row: get_warren_date(row["Date"]), axis=1)
```

```python
## Op Eds
#https://www.warren.senate.gov/index.cfm?p=op_eds&pg=2
def get_warren_oped_links():
    warren_oped_links = list()
    for i in range(1,6):
        url = "https://www.warren.senate.gov/index.cfm?p=op_eds&pg={}".format(str(i))
        driver.get(url)
        arr = driver.find_elements_by_class_name("title")
        for j in arr[:-6]:
            x = j.find_element_by_tag_name("a")
            warren_oped_links.append(x.get_attribute("href"))
    return warren_oped_links
    
warren_oped_links = get_warren_oped_links()
```

```python
driver.get("https://www.warren.senate.gov/?p=op_ed&id=455")

def get_warren_opeds(links):
    rows = list()
    for index, link in enumerate(links):
        print(index)
        row = list()
        driver.get(link)
        title = driver.find_element_by_class_name("page-title").text
        date = driver.find_element_by_class_name("inner-lyt-dates").text
        content = driver.find_element_by_class_name("record-bd")
        paras = content.find_elements_by_tag_name("p")
        release = ""
        for tag in paras:
            release += tag.text.strip()
        row = [date, title, release]
        rows.append(row)
    df = pd.DataFrame(rows, columns=["Date", "Title", "Release"])
    df = df[df.Release != ""]
    return df
            
    
    
warren_opeds_df = get_warren_opeds(warren_oped_links)
```

```python
def get_2014_candidates():
    page2014= "http://www.ontheissues.org/2014_Senate_Web.htm"
    driver.get(page2014)
    li_elems = driver.find_elements_by_tag_name("li")
    print(len(li_elems))
    names = list()
    for index, li in enumerate(li_elems):
        if "Senator" in li.text or "Challenger" in li.text:
            link = li.find_element_by_tag_name("a")
            names.append(link.text)
    return names
    
names14 = get_2014_candidates()
```

```python
def get_candidate_platforms(candidates, year):
    year = str(year)
    rows = list()
    for index, cand in enumerate(candidates):
        print(index)
        cand_url = cand.replace(" ", "_")
        url = "http://www.ontheissues.org/Archive/2014_Senate_Web_{}.htm".format(cand_url)
        driver.get(url)
        arr = driver.find_element_by_tag_name("body")
        platform = " ".join(arr.text.split("\n"))
        row = [int(year), cand, platform]
        rows.append(row)
    df = pd.DataFrame(rows, columns=["Year", "Candidate", "Platform"])
    return df
        
    
        
```
