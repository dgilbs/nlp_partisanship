### Small Business <a class="anchor" id="sbiz"></a>

```python
def get_sbc_archive_links_dems(party):
    links = list()
    for year in range(2010, 2015):
        print(year)
        url = "http://web.archive.org/web/20150110022702/http://www.sbc.senate.gov/public/index.cfm?p=DemocraticPressReleases&ContentType_id=4bfd610b-f7c6-4d07-9c74-7aab32dd9838&Group_id=92172669-d16f-4f1e-96dd-cc7f9b2a4d20&MonthDisplay=0&YearDisplay={}".format(str(year))
        driver.get(url)
        #print(url)
        arr = driver.find_elements_by_class_name("recordListTitle")
        for i in arr:
            link = i.find_element_by_tag_name("a")
            links.append(link.get_attribute("href"))
    return links
```

```python
dem_sb_links = get_sbc_archive_links_dems("Democratic")
```

```python
def get_sb_archive_releases(links, party):
    print(len(links))
    rows = list()
    for index, link in enumerate(links):
        print(index) 
        driver.get(link)
        
        try:
            date = driver.find_element_by_class_name("date").text
        except:
            date = ''
            
        try: 
            title = driver.find_element_by_class_name("title").text
        except:
            title = ''
            
        try:
            release = driver.find_element_by_class_name("content").text
        except:
            release = ''
        label = party + ", Small Business"
        row = [date, title, release, party, "Small Business", label]
        rows.append(row)
    df = pd.DataFrame(rows, columns=["Date", "Title", "Release", "Party", "Committee", "Label"])
    return df
```

```python
def get_old_sb_date(date_str):
    try:
        date_arr = date_str.split(" ")
        year= int(date_arr[-1])
        mo = date_arr[0]
        d = dict((v,k) for k,v in enumerate(calendar.month_abbr))
        month = d[mo]
        day = int(date_arr[1])
        return datetime(year=year, month=month, day=day)
    except:
        return date_str
    
```

```python
def get_month(date_obj):
    return date_obj.month

def get_year(date_obj):
    return date_obj.year

def get_day(date_obj):
    return date_obj.day
```

```python
dem_sb_df["Month"] = dem_sb_df.apply(lambda row: get_month(row["Date"]), axis=1)
dem_sb_df["Year"] = dem_sb_df.apply(lambda row: get_year(row["Date"]), axis=1)
dem_sb_df["Day"] = dem_sb_df.apply(lambda row: get_day(row["Date"]), axis=1)
```

## Foreign Relations <a class="anchor" id="frel"></a>

```python
def get_dem_fr_archives():
    links = list()
    for i in range(1,26):
        url = "http://web.archive.org/web/20121212021630/http://www.foreign.senate.gov/press/chair/?PageNum_rs={}&".format(str(i))
        driver.get(url)
        content = driver.find_element_by_id("tab-press-content")
        urls = content.find_elements_by_tag_name("a")
        for index, tag in enumerate(urls):
            links.append(tag.get_attribute("href"))
            
    return links
```

```python
def get_gop_fr_links():
    links = list()
    for i in range(1,10):
        url = "http://web.archive.org/web/20121211231428/http://www.foreign.senate.gov/press/ranking/?PageNum_rs={}&".format(str(i))
        driver.get(url)
        content = driver.find_element_by_id("tab-press-content")
        urls = content.find_elements_by_tag_name("a")
        for index, tag in enumerate(urls):
            links.append(tag.get_attribute("href"))
            
    return links
```

```python
dem_fr_links = get_dem_fr_archives()
```

```python
gop_fr_links = get_gop_fr_links()
```

```python
def get_fr_archives(links, party):
    print(len(links))
    rows = list()
    for index, link in enumerate(links):
        print(index)
        driver.get(link)
        try:
            date = driver.find_element_by_class_name("dateline").text
        except:
            date = ''
        try:
            title = driver.find_element_by_class_name("entry-title").text
        except:
            title = ''
        try:
            paras = driver.find_elements_by_tag_name("p")
            release = ""
            for tag in paras:
                release += tag.text
            release = release.split("# # #")[0]
        except:
            release = ''
        label = party + ", Foreign Relations"
        row = [date, title, release, party, "Foreign Relations", label]
        rows.append(row)
    df = pd.DataFrame(rows, columns=["Date", "Title", "Release", "Party", "Committee", "Label"])
    return df
```

```python
dem_fr_df = get_fr_archives(dem_fr_links, "Democratic")
```

```python
gop_fr_df = get_fr_archives(gop_fr_links, "Republican")
```
